Description: The 30-day police incident blotter data shows crimes that was reported on the block/intersection level.

```python
import pandas as pd 
```

```python
ReadIncident = pd.read_csv("data.csv") #import the data file 
ReadIncident.head(5) #print the first five rows
```

```python
NumOfIncident = ReadIncident.groupby(['INCIDENTNEIGHBORHOOD'])['INCIDENTNEIGHBORHOOD'].count() #group neighborhoods and count how many times each neighborhood shows up on the data
NumOfIncident.sort_values(ascending = True) #sort from least to greatest
```

```python
NumOfIncident.plot(figsize = (20, 10)) #plot the data and expanded the graph
```

```python
NumOfIncident.plot.bar(figsize = (180, 50)) # turn plot graph to bar graph and expanded the graph
```

Conclusion: East Carnegie is the safest neighborhood because only one incident report was made in the past 30 days.
